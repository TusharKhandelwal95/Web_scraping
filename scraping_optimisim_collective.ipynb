```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import csv
```

```python
URL = "https://gov.optimism.io/categories"
req = requests.get(URL)
```

```python
soup = BeautifulSoup(req.content, 'html.parser')
with open("web_page_content.csv",'w',encoding="utf-8") as f:
  f.write(str(soup))
```

```python
headlines = []
links = []
descriptions = []
```

```python
print(soup.prettify())
```

```python
categories = []
links = []
descriptions = []
```

```python
rows = soup.select('tbody > tr')

for row in rows:
    # Extract the category and link
    category_div = row.find('td', class_='category').find('div', itemprop='itemListElement')

    if category_div:
        # Extract the category text
        heading_tag = category_div.find('h3').find('span', itemprop='name')
        if heading_tag:
            category_text = heading_tag.text.strip()
            categories.append(category_text)

        # Extract the link
        link_tag = category_div.find('meta', itemprop='url')
        if link_tag:
            category_link = link_tag['content']
            links.append("https://gov.optimism.io/"+ category_link)

        # Extract the description
        description_tag = category_div.find('div', itemprop='description')
        description_text = description_tag.text.strip() if description_tag else 'No description available'
        descriptions.append(description_text)
```

```python
data = {
    'Categories': categories,
    'Links': links,
    'Descriptions': descriptions
}

df = pd.DataFrame(data)
df.to_csv("web_page_content.csv", index=False)
```
